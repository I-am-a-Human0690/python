```python
import matplotlib.pyplot as plt
student=[10,20,40,30,60]
course=["PHP","Java","Python","C#","Nodejs"]
```

```python
color=["purple","red","cyan","yellow","green"]
exp=[0,0,0,0.2,0.1]
plt.pie(student,labels=course,autopct="%.2f %%",colors=color,shadow=True,explode=exp,startangle=90)
plt.show()
```
