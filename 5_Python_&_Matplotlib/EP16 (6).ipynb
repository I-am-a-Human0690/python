```python
import matplotlib.pyplot as plt
product1=[10,20,30,40,50]
product2=[15,38,19,20,60]
month=[1,2,3,4,5]
```

```python
plt.plot(month,product1,"g+--")
plt.plot(month,product2,"r*-")
data={"size":20,"color":"red","backgroundcolor":"yellow"}
plt.xlabel("Month",data)
plt.ylabel("sales",data)
plt.title("Top Sale 2020",size=25,color="blue",loc="left")
plt.legend(["KK","OO"],loc=9,fontsize=15,facecolor="pink",edgecolor="green",title="THEPOND",title_fontsize="large")
plt.text(4.5,20,"ky",size="10",color="red")
plt.text(1.3,50,"ky",size="15",color="green")
plt.show()
```
